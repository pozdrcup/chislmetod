```python
import math

print("УПРАЖНЕНИЕ 1")
print("Параметры: β = 2, t = 2, P₁ = -1, P₂ = 1")

mantiss_on_exp = (2 - 1) * math.pow(2, 2 - 1)
print(f"1) Мантисс на одну экспоненту: (β-1) × β^(t-1) = {mantiss_on_exp}")

exp_count = 1 - (-1) + 1
print(f"2) Количество экспонент: P₂ - P₁ + 1 = {exp_count}")

positive_numbers = mantiss_on_exp * exp_count
print(f"3) Положительных чисел (без нуля): {positive_numbers}")

total_numbers = 2 * positive_numbers + 1
print(f"4) Общее количество чисел (с ± и нулём): {total_numbers}")

print("\nЧисла для числовой оси (подставляем в формулу x = ± m × β^e):")
print("  e = -1:  ±0.25, ±0.375")
print("  e =  0:  ±0.5,  ±0.75")
print("  e =  1:  ±1.0,  ±1.5")
print("  + 0.0")
print("Ось (13 точек):")
print("-1.5 ── -1.0 ── -0.75 ── -0.5 ── -0.375 ── -0.25 ── 0 ── 0.25 ── 0.375 ── 0.5 ── 0.75 ── 1.0 ── 1.5")

print("УПРАЖНЕНИЕ 2")
print("Параметры: β = 10, t = 3, P₁ = -9, P₂ = 9")

boundary = math.pow(10, -9 - 1)
print(f"Самое маленькое нормализованное число = β^(P₁-1)")
print(f"Границы машинного нуля: от -{boundary} до +{boundary}")
print("(всё, что |x| < 10^{-10}, автоматически становится 0)")
```
