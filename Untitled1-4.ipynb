```python
import math

# ================================================
# УПРАЖНЕНИЕ (одно большое задание из тетради)
# Часть 1: посчитать a) … и) с погрешностями
# Часть 2: написать функцию error_bounds
# ================================================

# Функция, которая по строке числа определяет его абсолютную погрешность
# (по последней значащей цифре, как в теории погрешностей)
def get_delta(num_str):
    # Заменяем запятую на точку, чтобы Python понял число
    if ',' in num_str:
        num_str = num_str.replace(',', '.')

    num = float(num_str)                    # само число

    # Считаем, сколько цифр после запятой
    if '.' in num_str:
        decimals = len(num_str.split('.')[1])
        unit = 10 ** (-decimals)            # единица последнего разряда
    else:
        unit = 1.0

    delta = 0.5 * unit                      # абсолютная погрешность = ±0.5 единицы
    return num, delta


# Список всех заданий с русскими буквами (точно как в тетради)
tasks = [
    ('а', '24,37', '9,18', '-'),     # а) 24,37 – 9,18
    ('б', '18,437', '24,9', '+'),    # б) 18,437 + 24,9
    ('в', '24,1', '0,037', '-'),     # в) 24,1 – 0,037
    ('г', '1,504', '1,502', '-'),    # г) 1,504 – 1,502
    ('д', '234,5', '194,3', '-'),    # д) 234,5 – 194,3
    ('е', '0,65', '1984', '*'),      # е) 0,65 · 1984
    ('ж', '12,6', '40,3', '*'),      # ж) 12,6 · 40,3
    ('з', '72,3', '0,34', '/'),      # з) 72,3 : 0,34
    ('и', '8124,6', '2,9', '/')      # и) 8124,6 : 2,9
]

print("=== УПРАЖНЕНИЯ: ПОГРЕШНОСТИ АРИФМЕТИЧЕСКИХ ДЕЙСТВИЙ ===\n")

# Перебираем каждое задание по очереди
for label, a_str, b_str, op in tasks:
    # Получаем число и его погрешность
    a, da = get_delta(a_str)
    b, db = get_delta(b_str)

    # Выполняем действие и считаем погрешности
    if op == '+':
        res = a + b
        d_res = da + db                          # при сложении/вычитании Δ = Δa + Δb
    elif op == '-':
        res = a - b
        d_res = da + db
    elif op == '*':
        res = a * b
        d_res = abs(res) * (da / abs(a) + db / abs(b))   # при умножении δ = δa + δb
    elif op == '/':
        res = a / b
        d_res = abs(res) * (da / abs(a) + db / abs(b))

    # Относительная погрешность
    delta_rel = d_res / abs(res) if abs(res) > 0 else float('inf')

    # Красивый вывод (точно как нужно в тетради)
    print(f"{label}) {a_str} {op} {b_str}")
    print(f"   Результат:          {res:.6f}")
    print(f"   Абсолютная погрешность Δ  = {d_res:.6f}")
    print(f"   Относительная погрешность δ = {delta_rel:.6f}  ({delta_rel*100:.4f} %)")
    print("   ---")


# ================================================
# ЧАСТЬ 2 — ФУНКЦИЯ, которую просили описать
# ================================================

print("\n=== ФУНКЦИЯ ДЛЯ ВЫЧИСЛЕНИЯ ГРАНИЦ ОТНОСИТЕЛЬНЫХ ПОГРЕШНОСТЕЙ ===")

def error_bounds(a, da, b, db, operation):
    """
    Универсальная функция для любых арифметических действий.
    a, b     — числа
    da, db   — их абсолютные погрешности
    operation — строка '+', '-', '*', '/'
    Возвращает: (результат, абсолютная погрешность, относительная погрешность)
    """
    if operation == '+' or operation == '-':
        result = a + b if operation == '+' else a - b
        abs_error = da + db                     # абсолютные погрешности складываются

    elif operation == '*':
        result = a * b
        abs_error = abs(result) * (da / abs(a) + db / abs(b)) if a != 0 and b != 0 else 0

    elif operation == '/':
        result = a / b
        abs_error = abs(result) * (da / abs(a) + db / abs(b)) if a != 0 and b != 0 else 0

    # Относительная погрешность
    rel_error = abs_error / abs(result) if abs(result) > 0 else float('inf')

    return result, abs_error, rel_error


# ================================================
# ДЕМОНСТРАЦИЯ ФУНКЦИИ error_bounds ДЛЯ ВСЕХ ЗАДАНИЙ
# (точно как просили в тетради — через функцию)
# ================================================

print("\n=== ДЕМОНСТРАЦИЯ ФУНКЦИИ error_bounds ДЛЯ ВСЕХ ЗАДАНИЙ а) — и) ===")
print("                (вызываем функцию error_bounds)\n")

# Перебираем ВСЕ задания из списка tasks
for label, a_str, b_str, op in tasks:
    # Получаем числа и их абсолютные погрешности
    a, da = get_delta(a_str)
    b, db = get_delta(b_str)

    # Вызываем ТВОЮ функцию
    result, abs_err, rel_err = error_bounds(a, da, b, db, op)

    # Красивый вывод — точно как на скриншоте
    print(f"{label}) {a_str} {op} {b_str}")
    print(f"   Результат:          {result:.6f}")
    print(f"   Абсолютная погрешность Δ  = {abs_err:.6f}")
    print(f"   Относительная погрешность δ = {rel_err:.6f}  ({rel_err*100:.4f} %)")
    print("   ---")
```
